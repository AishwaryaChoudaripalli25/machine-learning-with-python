```python
# filter fn using lambda
nums=[1,2,4,5,6,7,3,2,6,8,9]
evens=list(filter(lambda n: n%2==0,nums))
print(evens)
```

```python
#map fn without lambda
def update(n):
    return n*3
nums=[1,2,4,5,6,7,3,2,6,8,9]
evens=list(filter(lambda n: n%2==0,nums))
print(evens)
triples=list(map(update,evens))
print(triples)
```

```python
#map fn with lambda
nums=[1,2,4,5,6,7,3,2,6,8,9]
evens=list(filter(lambda n: n%2==0,nums))
print(evens)
triples=list(map(lambda n:n*3,evens))
print(triples)
```

```python
#reduce fn without lambda
from functools import reduce
def multiply_all(a,b):
    return a*b
nums=[1,2,4,5,6,7,3,2,6,8,9]
odds=list(filter(lambda n: n%2!=0,nums))
print(odds)
triples=list(map(lambda n:n*3,odds))
print(triples)
multi=reduce(multiply_all,triples)
print(multi)
mult=reduce(multiply_all,nums)
print(mult)
```

```python
#reduce fn with lambda
nums=[1,2,4,5,6,7,3,2,6,8,9]
odds=list(filter(lambda n: n%2!=0,nums))
print(odds)
triples=list(map(lambda n:n*3,odds))
print(triples)
multi=reduce(lambda a,b:a*b,triples)
print(multi)
```

```python
# factorial
def factorial(x):
    k=1
    for i in range(1,x+1):
        k=k*i
    return k
```

```python
class student:
    def __init__(self,dl,befa):
        self.dl=dl
        self.befa=befa
    def __gt__(self,other):
        g1=self.dl+self.befa
        g2=other.dl+other.befa
        if g1>g2:
            print("g1 is greater")
        else:
            print("g2 is greater")
s1=student(20,30)
s2=student(30,40)
s3=student.__gt__(s1,s2)
```

```python
class student:
    def __init__(self,dl,befa):
        self.dl=dl
        self.befa=befa
    def __ge__(self,other):
        g1=self.dl+self.befa
        g2=other.dl+other.befa
        if g1>g2:
            print("g1 is greater")
        elif g1==g2:
            print("g1 is equal to g2")
        else:
            print("g2 is greater")
s1=student(20,30)
s2=student(30,20)
s3=student.__ge__(s1,s2)
```

## method overloading

```python

class sum():
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def add(a,b):
        c=a+b
        return c
    def add(a,b,c):
        d=a+b+c
        return d
k=sum.add(2,3)
print(k)
        
    
```

```python
class sum():
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def add(a=None,b=None,c=None):
        if a!=None and b!=None and c!=None:
            d=a+b+c
        elif a!=None and b!=None:
            d=a+b
        else:
            d=a
        return d
k=sum.add(2,3)
print(k)
        
```

```python
#function to take multiple arguments
class data:
    def __init__(self,datatype,ans):
        self.datatype=datatype
        self.ans=ans
    def add(self,datatype,ans,*args):
        if datatype=='int':
            ans=0
        if datatype=='str':
            ans=''
        for x in args:
            ans=ans+x
        return ans
data.add('int',3,9)
```
